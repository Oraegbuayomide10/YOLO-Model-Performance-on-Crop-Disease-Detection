Set Paths and Load CSV

```python
# Set the data directory
ROOT_DIR  = Path(os.getcwd())
SUBS_DIR = ROOT_DIR / 'subs'
VAL_IMAGES_DIR = ROOT_DIR / r'val\images\fold_0' # path to where val images for fold_0 are saved

# load predictions csvs for both YOLO-11 and RT-DETR 
detr = pd.read_csv(SUBS_DIR / 'RTDETR_submission.csv') 
yolo = pd.read_csv(SUBS_DIR / 'YOLO_submission.csv')
validation = pd.read_csv(ROOT_DIR / 'non_duplicates_train.csv') 
```

Configuration

```python
class CFG:
    num_images = 3
```

Visualize Predictions from both Models

```python
# pick random images
all_val_images = os.listdir(VAL_IMAGES_DIR)
random_image_ids = [all_val_images[np.random.randint(0, len(all_val_images))] for id in range(CFG.num_images)]

plot_images_against(selected_images = random_image_ids,
                    dataframe_1=yolo,
                    dataframe_2=detr,
                    dataframe_original=validation ,
                    images_fold_dir=VAL_IMAGES_DIR,
                    dataframe_1_title='YOLO',
                    dataframe_2_title='RT-DETR',
                    )
```
